```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
E = np.loadtxt("E.txt")
M = np.loadtxt("M.txt")
C = np.loadtxt("C.txt")
T=np.linspace(2.5,1.5,21)
```

```python
plt.plot(T,E,'.')
plt.xlabel('T')
plt.ylabel('E')
plt.figure()
plt.plot(T,M,'.')
plt.xlabel('T')
plt.ylabel('M')
plt.figure()
plt.plot(T,C,'.')
plt.xlabel('T')
plt.ylabel('C')
plt.figure()
```
